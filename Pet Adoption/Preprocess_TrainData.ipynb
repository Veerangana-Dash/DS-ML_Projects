# Preprocessing module for Pet Adoption :

#### Import libraries :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

#### Load data :


```python
dataset = pd.read_csv('train.csv')
dataset
```

#### Checkpoint :

```python
df = dataset.copy()
```

#### Summary :

```python
df.info()
```

#### Drop Pet_ID :

```python
df = df.drop(['pet_id'],axis=1)
df.head()
```

#### Analyzing variables/features :

```python
df['Date Difference'] = pd.to_datetime(df['listing_date']) - pd.to_datetime(df['issue_date'])
df['Date Difference'] = (df['Date Difference'] / np.timedelta64(1,'D')).round()
df.head()
```

```python
#checkpoint
df = df.drop(['issue_date'],axis=1)
df= df.drop(['listing_date'],axis =1)

df_dated = df.copy()
df_dated.head()
```

```python
df_dated.columns.values
```

```python
column_names_reordered = ['Date Difference','condition', 'color_type', 'length(m)', 'height(cm)', 'X1', 'X2','breed_category', 'pet_category'] 
df_dated = df_dated[column_names_reordered]
df_dated.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df_colored = df_dated.copy()
df_colored['color_encoded'] = LabelEncoder().fit_transform(df_colored['color_type'])
#df_colored = df_colored[['Date Difference','condition','color_encoded','length(m)','height(cm)','X1','X2','breed_category', 'pet_category']].fillna(-1)
df_colored
```

```python
df_old = df_dated['color_type'].value_counts().plot(kind='barh')
df_old.plot(figsize=(20,20))
```

```python
df_colored['Size'] = df_colored['length(m)'] * df_colored['height(cm)'] * 0.01 
df_colored.head()
```

```python
df_colored = df_colored.drop(['length(m)' , 'height(cm)','color_type'], axis=1)
df_colored.head()
```

```python
df_colored['condition'] = df_colored['condition'].fillna(round(df_colored['condition'].mean()))
df_colored.head()
```

```python
df_colored.columns.values
```

```python
column_names_reordered = ['Date Difference', 'condition', 'color_encoded', 'Size','X1', 'X2', 'breed_category',
       'pet_category']
df_ordered = df_colored[column_names_reordered]
df_ordered.head()
```

```python
df_ordered.shape
```

```python
df_ordered.to_csv('processed_train_data.csv',index=False)
```
